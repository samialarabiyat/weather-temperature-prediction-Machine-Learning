```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# 3. Prepare the sequential data using only the temperature and timestamp
df_lstm = df[['Formatted Date', 'Temperature (C)']].copy()
df_lstm['Formatted Date'] = pd.to_datetime(df_lstm['Formatted Date'])
df_lstm = df_lstm.sort_values('Formatted Date').reset_index(drop=True)

# 4. Normalize temperature values to be between 0 and 1
scaler = MinMaxScaler()
temp_scaled = scaler.fit_transform(df_lstm[['Temperature (C)']])
```

```python
# 5. Create sliding window sequences
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

window_size = 24  # e.g., using 24 hours of data to predict the next hour
X_seq, y_seq = create_sequences(temp_scaled, window_size)
```

```python
# 6. Split data into training and test sets
split_index = int(len(X_seq) * 0.8)
X_train, X_test = X_seq[:split_index], X_seq[split_index:]
y_train, y_test = y_seq[:split_index], y_seq[split_index:]

# 7. Build the LSTM model
model = Sequential([
    LSTM(64, input_shape=(window_size, 1), return_sequences=False),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
# 8. Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_split=0.1)

# 9. Predict on test data
y_pred_lstm = model.predict(X_test)
```

```python
# 10. Inverse transform the scaled data to get actual temperature values
y_test_rescaled = scaler.inverse_transform(y_test)
y_pred_rescaled = scaler.inverse_transform(y_pred_lstm)
```

```python
import matplotlib.pyplot as plt
# 11. Plot actual vs predicted temperature
plt.figure(figsize=(10,5))
plt.plot(y_test_rescaled, label="Actual")
plt.plot(y_pred_rescaled, label="LSTM Prediction")
plt.title("LSTM - Temperature Forecast")
plt.xlabel("Time Step")
plt.ylabel("Temperature (C)")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Regression Metrics
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

print("📊 LSTM Regression Metrics:")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"R² Score: {r2:.4f}")

# 2. Custom Accuracy (within ±2°C)
threshold = 2.0
within_range = np.abs(y_test_rescaled - y_pred_rescaled) <= threshold
accuracy = np.mean(within_range)

print(f"\n✅ Custom Accuracy (±{threshold}°C): {accuracy * 100:.2f}%")
```
